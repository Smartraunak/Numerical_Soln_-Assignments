```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lax_fedrich(N_x,ratio,T_unit,function):
    if (function=='linear'):
        delta_x=(2)/(N_x-1)
        delta_t=ratio*delta_x
        N_t=int((T_unit/delta_t)+1)
        data=np.zeros((N_x,N_t))
        for i in range(N_x):
            if (abs(-1+i*delta_x)<1/3):
                data[i][0]=1
            else:
                data[i][0]=0
        for j in range(N_t-1):
            for i in range(N_x-2):
                data[i+1][j+1]=(1/2)*(data[i+2][j]+data[i][j])-(ratio/2)*(data[i+2][j]-data[i][j])
            data[0][j+1]=data[N_x-2][j+1]
            data[N_x-1][j+1]=data[1][j+1]
        x=np.linspace(-1,1,N_x)
        y=np.linspace(0,T_unit,N_t)
        plt.plot(x,data[:,int(4/delta_t)], label= f'N = {N_x}')
        plt.xlabel('x')
        plt.ylabel('u(x)')
        plt.title('Hyperboilx PDE (f(x)=x)')
        plt.show()
    elif (function=='quadratic'):
        delta_x=(2)/(N_x-1)
        delta_t=ratio*delta_x
        N_t=int((T_unit/delta_t)+1)
        data=np.zeros((N_x,N_t))
        for i in range(N_x):
            if (abs(-1+i*delta_x)<1/3):
                data[i][0]=1
            else:
                data[i][0]=-1
        for j in range(N_t-1):
            for i in range(N_x-2):
                data[i+1][j+1]=(1/2)*(data[i+2][j]+data[i][j])-(ratio/4)*(np.square(data[i+2][j])-np.square(data[i][j]))
            data[0][j+1]=data[N_x-2][j+1]
            data[N_x-1][j+1]=data[1][j+1]
        x=np.linspace(-1,1,N_x)
        y=np.linspace(0,T_unit,N_t)
        plt.plot(x,data[:,int(0.3/delta_t)], label= f'N = {N_x}')
        plt.xlabel('x')
        plt.ylabel('u(x)')
        plt.title('Hyperboilx PDE (f(x)=x^2/2)')
        plt.show()
```

```python
lax_fedrich(201,0.8,4,'linear')
lax_fedrich(5001,0.8,4,'linear')
```

```python
lax_fedrich(201,0.8,4,'quadratic')
lax_fedrich(5001,0.8,4,'quadratic')
```
